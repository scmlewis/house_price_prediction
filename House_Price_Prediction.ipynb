# Step 1: EDA

```python
# %% [markdown]
# # ISOM5610: Introduction to Business Analytics
# ## Step 1: Data Understanding and Exploratory Data Analysis (EDA)
# This notebook performs EDA on the Ames house price dataset to understand its structure, identify patterns, and detect issues (e.g., missing values, outliers, multicollinearity). Outputs are displayed inline for review.

# %% [markdown]
# ### Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.outliers_influence import variance_inflation_factor
from IPython.display import display

# Set up plotting style
try:
    plt.style.use('ggplot')  # Use 'ggplot' style for clean aesthetics
except OSError:
    print("Warning: 'ggplot' style not found. Using default Matplotlib style.")
sns.set(font_scale=1.2)

# %% [markdown]
# ### Step 1.1: Load the Dataset
def load_data(file_path='data/Ames.csv'):
    try:
        df = pd.read_csv(file_path)
        print(f"Dataset loaded with {df.shape[0]} rows and {df.shape[1]} columns.")
        return df
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found. Please ensure 'Ames.csv' is in the 'data' subfolder.")
        raise

# Load dataset
df = load_data()

# %% [markdown]
# ### Step 1.2: Summarize Data
def summarize_data(df):
    # Descriptive statistics for numerical variables
    numerical_summary = df.select_dtypes(include=['int64', 'float64']).describe()
    print("\nDescriptive Statistics for Numerical Variables:")
    display(numerical_summary)
    
    # Summary for categorical variables
    categorical_cols = df.select_dtypes(include=['object']).columns
    print("\nCategorical Variables Summary:")
    for col in categorical_cols:
        print(f"{col}: {df[col].nunique()} unique values, Mode: {df[col].mode()[0]}")
    
    return numerical_summary, categorical_cols

# Summarize data
numerical_summary, categorical_cols = summarize_data(df)

# %% [markdown]
# ### Step 1.3: Visualize Distributions
def plot_distributions(df, key_numerical=['SalePrice', 'GrLivArea', 'LotArea', 'TotalBsmtSF'], 
                       key_categorical=['Neighborhood', 'OverallQual']):
    # Numerical variables: Histograms and Boxplots
    for col in key_numerical:
        plt.figure(figsize=(12, 5))
        
        # Histogram
        plt.subplot(1, 2, 1)
        sns.histplot(df[col], kde=True)
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        
        # Boxplot
        plt.subplot(1, 2, 2)
        sns.boxplot(y=df[col])
        plt.title(f'Boxplot of {col}')
        
        plt.tight_layout()
        plt.show()
    
    # Categorical variables: Boxplots of SalePrice
    for col in key_categorical:
        plt.figure(figsize=(10, 6))
        sns.boxplot(x=col, y='SalePrice', data=df)
        plt.title(f'SalePrice by {col}')
        plt.xticks(rotation=45)
        plt.show()

# Plot distributions
plot_distributions(df)

# %% [markdown]
# ### Step 1.4: Explore Relationships
def explore_relationships(df, predictors=['GrLivArea', 'LotArea', 'TotalBsmtSF']):
    # Scatterplots
    for col in predictors:
        plt.figure(figsize=(8, 6))
        sns.scatterplot(x=df[col], y=df['SalePrice'])
        plt.title(f'SalePrice vs {col}')
        plt.xlabel(col)
        plt.ylabel('SalePrice')
        plt.show()
    
    # Correlation matrix
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
    corr_matrix = df[numerical_cols].corr()
    plt.figure(figsize=(12, 10))
    sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', center=0)
    plt.title('Correlation Matrix')
    plt.show()
    
    # Top correlations with SalePrice
    saleprice_corr = corr_matrix['SalePrice'].sort_values(ascending=False)
    print("\nTop Correlations with SalePrice:")
    display(saleprice_corr.head(10))
    return corr_matrix

# Explore relationships
corr_matrix = explore_relationships(df)

# %% [markdown]
# ### Step 1.5: Check for Multicollinearity
def check_multicollinearity(df, predictors=['GrLivArea', 'LotArea', 'TotalBsmtSF', 'OverallQual', 
                                            'YearBuilt', '1stFlrSF']):
    # Prepare data for VIF
    X = df[predictors].fillna(df[predictors].median())
    X = (X - X.mean()) / X.std()  # Standardize to avoid scale issues
    vif_data = pd.DataFrame()
    vif_data['Predictor'] = predictors
    vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    print("\nVariance Inflation Factors:")
    display(vif_data)
    return vif_data

# Check multicollinearity
vif_data = check_multicollinearity(df)

# %% [markdown]
# ### Step 1.6: Assess Missing Values
def assess_missing_values(df):
    missing = df.isnull().sum()
    missing_percent = (missing / len(df)) * 100
    missing_df = pd.DataFrame({'Missing Count': missing, 'Missing Percent': missing_percent})
    missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values(by='Missing Count', ascending=False)
    print("\nMissing Values:")
    display(missing_df)
    return missing_df

# Assess missing values
missing_df = assess_missing_values(df)

# %% [markdown]
# ### Step 1.7: Detect Outliers
def detect_outliers(df, columns=['SalePrice', 'GrLivArea', 'LotArea']):
    outliers = {}
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outlier_count = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
        outliers[col] = outlier_count
        print(f"\nOutliers in {col}: {outlier_count}")
    return outliers

# Detect outliers
outliers = detect_outliers(df)
```

# Step 2: Data Preprocessing

```python
# %% [markdown]
# # ISOM5610: Introduction to Business Analytics
# ## Step 2: Data Preprocessing
# This notebook preprocesses the Ames house price dataset by handling missing values, encoding categorical variables, transforming skewed variables, engineering features, and partitioning the data for modeling. Outputs are displayed inline for review.

# %% [markdown]
# ### Import Libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from IPython.display import display

# Set random seed for reproducibility
np.random.seed(42)

# %% [markdown]
# ### Load Dataset
def load_data(file_path='data/Ames.csv'):
    try:
        df = pd.read_csv(file_path)
        print(f"Dataset loaded with {df.shape[0]} rows and {df.shape[1]} columns.")
        return df
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found. Please ensure 'Ames.csv' is in the 'data' subfolder.")
        raise

df = load_data()

# %% [markdown]
# ### Preprocess Data
def preprocess_data(df):
    # 1. Handle Missing Values
    # Numerical: Impute with median
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
    for col in numerical_cols:
        df[col].fillna(df[col].median(), inplace=True)
    
    # Categorical: Replace with "None" or mode
    none_cols = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 
                 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 
                 'BsmtFinType2', 'GarageType', 'GarageFinish', 'GarageQual', 
                 'GarageCond', 'MasVnrType']
    for col in none_cols:
        df[col].fillna('None', inplace=True)
    df['Electrical'].fillna(df['Electrical'].mode()[0], inplace=True)
    
    # Drop columns with high missingness
    drop_cols = ['PoolQC', 'MiscFeature', 'Alley', 'Fence']
    df = df.drop(columns=drop_cols)
    print(f"Dropped columns: {drop_cols}")
    
    # 2. Encode Categorical Variables
    # Ordinal encoding for quality-related variables
    quality_map = {'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5, 'None': 0}
    ordinal_cols = ['ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond', 
                    'HeatingQC', 'KitchenQual', 'FireplaceQu', 'GarageQual', 
                    'GarageCond']
    for col in ordinal_cols:
        df[col] = df[col].map(quality_map, na_action='ignore')
    
    # Dummy encoding for nominal categorical variables
    categorical_cols = ['Neighborhood', 'MSZoning', 'HouseStyle', 'BldgType', 
                       'GarageType', 'SaleCondition']
    df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
    print(f"Encoded categorical variables: {categorical_cols}")
    
    # 3. Transform Variables
    # Log-transform skewed variables
    skewed_cols = ['SalePrice', 'GrLivArea', 'LotArea', 'TotalBsmtSF']
    for col in skewed_cols:
        df[f'Log_{col}'] = np.log1p(df[col])  # log1p handles zero values
    
    # 4. Feature Engineering
    df['TotalSF'] = df['TotalBsmtSF'] + df['1stFlrSF'] + df['2ndFlrSF']
    df['Age'] = df['YrSold'] - df['YearBuilt']
    df['Qual_GrLivArea'] = df['OverallQual'] * df['Log_GrLivArea']
    print("Engineered features: TotalSF, Age, Qual_GrLivArea")
    
    # 5. Standardize Numerical Predictors
    scaler = StandardScaler()
    num_cols = ['Log_GrLivArea', 'Log_LotArea', 'Log_TotalBsmtSF', 'TotalSF', 
                'Age', 'OverallQual', 'Qual_GrLivArea']
    df[num_cols] = scaler.fit_transform(df[num_cols])
    print("Standardized numerical predictors:", num_cols)
    
    # Display processed dataset summary
    print("\nProcessed Dataset Summary:")
    display(df[num_cols + ['Log_SalePrice']].describe())
    
    return df

# Preprocess data
df_processed = preprocess_data(df)
print(f"Preprocessed dataset shape: {df_processed.shape}")

# %% [markdown]
# ### Data Partition
def partition_data(df, target='Log_SalePrice', test_size=0.3):
    # Select features (exclude original SalePrice and ID)
    features = [col for col in df.columns if col not in ['SalePrice', 'Log_SalePrice', 'Id']]
    X = df[features]
    y = df[target]
    
    # Split into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)
    print(f"Training set shape: {X_train.shape}, Testing set shape: {X_test.shape}")
    
    return X_train, X_test, y_train, y_test

# Partition data
X_train, X_test, y_train, y_test = partition_data(df_processed)

# %% [markdown]
# ### Save Processed Data (Optional)
# Optionally save processed data for modeling
# df_processed.to_csv('data/ames_processed.csv', index=False)
# print("Processed data saved to 'data/ames_processed.csv'")
```

# Step 3: Model Building

```python
# %% [markdown]
# # ISOM5610: Introduction to Business Analytics
# ## Step 3: Model Building
# This notebook builds and compares regression models to predict Log_SalePrice using the preprocessed Ames dataset. Models include multiple linear regression, quadratic regression, and LASSO regression. Outputs are displayed inline.

# %% [markdown]
# ### Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.metrics import mean_squared_error
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.stats.diagnostic import het_breuschpagan
from scipy.stats import shapiro
import statsmodels.api as sm
from IPython.display import display

# Set random seed for reproducibility
np.random.seed(42)

# Set plotting style
try:
    plt.style.use('ggplot')
except OSError:
    print("Warning: 'ggplot' style not found. Using default Matplotlib style.")
sns.set(font_scale=1.2)

# %% [markdown]
# ### Load and Preprocess Data
def load_and_preprocess_data(file_path='data/Ames.csv'):
    try:
        df = pd.read_csv(file_path)
        print(f"Dataset loaded with {df.shape[0]} rows and {df.shape[1]} columns.")
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found. Please ensure 'Ames.csv' is in the 'data' subfolder.")
        raise
    
    # Handle missing values
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
    for col in numerical_cols:
        df[col] = df[col].fillna(df[col].median())
    none_cols = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 
                 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 
                 'BsmtFinType2', 'GarageType', 'GarageFinish', 'GarageQual', 
                 'GarageCond', 'MasVnrType']
    for col in none_cols:
        df[col] = df[col].fillna('None')
    df['Electrical'] = df['Electrical'].fillna(df['Electrical'].mode()[0])
    df = df.drop(columns=['PoolQC', 'MiscFeature', 'Alley', 'Fence'])
    
    # Encode categorical variables
    # Expanded quality_map to handle all possible values
    quality_map = {'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5, 'None': 0, 
                   np.nan: 0, 'Unknown': 0}  # Added fallback for unexpected values
    ordinal_cols = ['ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond', 
                    'HeatingQC', 'KitchenQual', 'FireplaceQu', 'GarageQual', 
                    'GarageCond']
    for col in ordinal_cols:
        df[col] = df[col].map(quality_map, na_action='ignore').fillna(0).astype(float)
    
    # Dummy encoding for all nominal categorical variables
    categorical_cols = ['MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities', 
                        'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 
                        'Condition2', 'BldgType', 'HouseStyle', 'RoofStyle', 
                        'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 
                        'Foundation', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 
                        'Heating', 'CentralAir', 'Electrical', 'Functional', 
                        'GarageType', 'GarageFinish', 'PavedDrive', 'SaleType', 
                        'SaleCondition']
    df = pd.get_dummies(df, columns=categorical_cols, drop_first=True, dtype=float)
    print(f"Encoded categorical variables: {categorical_cols}")
    
    # Log-transform skewed variables
    skewed_cols = ['SalePrice', 'GrLivArea', 'LotArea', 'TotalBsmtSF']
    for col in skewed_cols:
        df[f'Log_{col}'] = np.log1p(df[col]).astype(float)
    
    # Feature engineering
    df['TotalSF'] = (df['TotalBsmtSF'] + df['1stFlrSF'] + df['2ndFlrSF']).astype(float)
    df['Age'] = (df['YrSold'] - df['YearBuilt']).astype(float)
    df['Qual_GrLivArea'] = (df['OverallQual'] * df['Log_GrLivArea']).astype(float)
    print("Engineered features: TotalSF, Age, Qual_GrLivArea")
    
    # Standardize numerical predictors
    scaler = StandardScaler()
    num_cols = ['Log_GrLivArea', 'Log_LotArea', 'Log_TotalBsmtSF', 'TotalSF', 
                'Age', 'OverallQual', 'Qual_GrLivArea']
    df[num_cols] = scaler.fit_transform(df[num_cols]).astype(float)
    print("Standardized numerical predictors:", num_cols)
    
    # Verify all columns are numeric
    non_numeric_cols = df.select_dtypes(include=['object']).columns
    if len(non_numeric_cols) > 0:
        print(f"Warning: Non-numeric columns detected: {non_numeric_cols}")
        raise ValueError("Non-numeric columns found after preprocessing.")
    else:
        print("All columns are numeric.")
    
    # Display processed dataset summary
    print("\nProcessed Dataset Summary:")
    display(df[['Log_SalePrice', 'Log_GrLivArea', 'Log_LotArea', 'Log_TotalBsmtSF', 
                'TotalSF', 'Age', 'OverallQual', 'Qual_GrLivArea']].describe())
    
    return df

# Load and preprocess
df = load_and_preprocess_data()
print(f"Preprocessed dataset shape: {df.shape}")

# %% [markdown]
# ### Data Partition
def partition_data(df, target='Log_SalePrice'):
    features = [col for col in df.columns if col not in ['SalePrice', 'Log_SalePrice', 'Id']]
    X = df[features]
    y = df[target].astype(float)
    
    # Verify X is numeric
    if X.select_dtypes(include=['object']).columns.any():
        raise ValueError(f"Non-numeric columns in X: {X.select_dtypes(include=['object']).columns}")
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    print(f"Training set shape: {X_train.shape}, Testing set shape: {X_test.shape}")
    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = partition_data(df)

# %% [markdown]
# ### Model 1: Multiple Linear Regression
def fit_linear_regression(X_train, X_test, y_train, y_test):
    # Fit model
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    # Predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    # Performance metrics
    train_mse = mean_squared_error(y_train, y_train_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)
    train_rmse = np.sqrt(train_mse)
    test_rmse = np.sqrt(test_mse)
    r2_train = model.score(X_train, y_train)
    r2_test = model.score(X_test, y_test)
    
    # Statsmodels for detailed summary
    X_train_sm = sm.add_constant(X_train)
    model_sm = sm.OLS(y_train, X_train_sm).fit()
    
    # Residual analysis
    residuals = y_train - y_train_pred
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    sns.scatterplot(x=y_train_pred, y=residuals)
    plt.axhline(0, color='red', linestyle='--')
    plt.title('Residuals vs Fitted (Linear Regression)')
    plt.xlabel('Fitted Values')
    plt.ylabel('Residuals')
    
    plt.subplot(1, 2, 2)
    sm.qqplot(residuals, line='45', fit=True)
    plt.title('Q-Q Plot (Linear Regression)')
    plt.tight_layout()
    plt.show()
    
    # Breusch-Pagan test for heteroskedasticity
    _, pval_bp, _, _ = het_breuschpagan(residuals, X_train_sm)
    # Shapiro-Wilk test for normality
    _, pval_sw = shapiro(residuals)
    
    print("\nLinear Regression Results:")
    print(f"Train MSE: {train_mse:.4f}, Train RMSE: {train_rmse:.4f}")
    print(f"Test MSE: {test_mse:.4f}, Test RMSE: {test_rmse:.4f}")
    print(f"Train R-squared: {r2_train:.4f}, Test R-squared: {r2_test:.4f}")
    print(f"Breusch-Pagan p-value: {pval_bp:.4f} (H0: Homoskedasticity)")
    print(f"Shapiro-Wilk p-value: {pval_sw:.4f} (H0: Normality)")
    print("\nModel Summary:")
    display(model_sm.summary())
    
    return model, model_sm, train_mse, test_mse

# Fit linear regression
lin_reg, lin_reg_sm, lin_train_mse, lin_test_mse = fit_linear_regression(X_train, X_test, y_train, y_test)

# %% [markdown]
# ### Model 2: Quadratic Regression
def fit_quadratic_regression(X_train, X_test, y_train, y_test):
    # Add squared terms for key predictors
    X_train_quad = X_train.copy()
    X_test_quad = X_test.copy()
    quad_cols = ['Log_GrLivArea', 'Log_TotalBsmtSF']
    for col in quad_cols:
        X_train_quad[f'{col}_squared'] = X_train_quad[col] ** 2
        X_test_quad[f'{col}_squared'] = X_test_quad[col] ** 2
    
    # Fit model
    model = LinearRegression()
    model.fit(X_train_quad, y_train)
    
    # Predictions
    y_train_pred = model.predict(X_train_quad)
    y_test_pred = model.predict(X_test_quad)
    
    # Performance metrics
    train_mse = mean_squared_error(y_train, y_train_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)
    train_rmse = np.sqrt(train_mse)
    test_rmse = np.sqrt(test_mse)
    r2_train = model.score(X_train_quad, y_train)
    r2_test = model.score(X_test_quad, y_test)
    
    # Residual analysis
    residuals = y_train - y_train_pred
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    sns.scatterplot(x=y_train_pred, y=residuals)
    plt.axhline(0, color='red', linestyle='--')
    plt.title('Residuals vs Fitted (Quadratic Regression)')
    plt.xlabel('Fitted Values')
    plt.ylabel('Residuals')
    
    plt.subplot(1, 2, 2)
    sm.qqplot(residuals, line='45', fit=True)
    plt.title('Q-Q Plot (Quadratic Regression)')
    plt.tight_layout()
    plt.show()
    
    # Breusch-Pagan test
    X_train_sm = sm.add_constant(X_train_quad)
    _, pval_bp, _, _ = het_breuschpagan(residuals, X_train_sm)
    # Shapiro-Wilk test
    _, pval_sw = shapiro(residuals)
    
    print("\nQuadratic Regression Results:")
    print(f"Train MSE: {train_mse:.4f}, Train RMSE: {train_rmse:.4f}")
    print(f"Test MSE: {test_mse:.4f}, Test RMSE: {test_rmse:.4f}")
    print(f"Train R-squared: {r2_train:.4f}, Test R-squared: {r2_test:.4f}")
    print(f"Breusch-Pagan p-value: {pval_bp:.4f} (H0: Homoskedasticity)")
    print(f"Shapiro-Wilk p-value: {pval_sw:.4f} (H0: Normality)")
    
    return model, train_mse, test_mse

# Fit quadratic regression
quad_reg, quad_train_mse, quad_test_mse = fit_quadratic_regression(X_train, X_test, y_train, y_test)

# %% [markdown]
# ### Model 3: LASSO Regression
def fit_lasso_regression(X_train, X_test, y_train, y_test):
    # Fit LASSO with cross-validation to select alpha
    alphas = np.logspace(-4, 1, 50)
    best_alpha = None
    best_score = -np.inf
    for alpha in alphas:
        model = Lasso(alpha=alpha, max_iter=10000)
        scores = cross_val_score(model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
        if scores.mean() > best_score:
            best_score = scores.mean()
            best_alpha = alpha
    
    # Fit final model
    model = Lasso(alpha=best_alpha, max_iter=10000)
    model.fit(X_train, y_train)
    
    # Predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    # Performance metrics
    train_mse = mean_squared_error(y_train, y_train_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)
    train_rmse = np.sqrt(train_mse)
    test_rmse = np.sqrt(test_mse)
    r2_train = model.score(X_train, y_train)
    r2_test = model.score(X_test, y_test)
    
    # Residual analysis
    residuals = y_train - y_train_pred
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    sns.scatterplot(x=y_train_pred, y=residuals)
    plt.axhline(0, color='red', linestyle='--')
    plt.title('Residuals vs Fitted (LASSO Regression)')
    plt.xlabel('Fitted Values')
    plt.ylabel('Residuals')
    
    plt.subplot(1, 2, 2)
    sm.qqplot(residuals, line='45', fit=True)
    plt.title('Q-Q Plot (LASSO Regression)')
    plt.tight_layout()
    plt.show()
    
    # Breusch-Pagan test
    X_train_sm = sm.add_constant(X_train)
    _, pval_bp, _, _ = het_breuschpagan(residuals, X_train_sm)
    # Shapiro-Wilk test
    _, pval_sw = shapiro(residuals)
    
    # Selected features
    selected_features = X_train.columns[model.coef_ != 0].tolist()
    
    print("\nLASSO Regression Results:")
    print(f"Best alpha: {best_alpha:.6f}")
    print(f"Train MSE: {train_mse:.4f}, Train RMSE: {train_rmse:.4f}")
    print(f"Test MSE: {test_mse:.4f}, Test RMSE: {test_rmse:.4f}")
    print(f"Train R-squared: {r2_train:.4f}, Test R-squared: {r2_test:.4f}")
    print(f"Breusch-Pagan p-value: {pval_bp:.4f} (H0: Homoskedasticity)")
    print(f"Shapiro-Wilk p-value: {pval_sw:.4f} (H0: Normality)")
    print(f"Selected features: {selected_features}")
    
    return model, train_mse, test_mse, selected_features

# Fit LASSO regression
lasso_reg, lasso_train_mse, lasso_test_mse, selected_features = fit_lasso_regression(X_train, X_test, y_train, y_test)

# %% [markdown]
# ### Model Comparison
def compare_models(models_results):
    comparison = pd.DataFrame({
        'Model': ['Linear Regression', 'Quadratic Regression', 'LASSO Regression'],
        'Train MSE': [lin_train_mse, quad_train_mse, lasso_train_mse],
        'Test MSE': [lin_test_mse, quad_test_mse, lasso_test_mse],
        'Train RMSE': [np.sqrt(lin_train_mse), np.sqrt(quad_train_mse), np.sqrt(lasso_train_mse)],
        'Test RMSE': [np.sqrt(lin_test_mse), np.sqrt(quad_test_mse), np.sqrt(lasso_test_mse)]
    })
    print("\nModel Comparison:")
    display(comparison)
    
    # Select best model based on Test MSE
    best_model = comparison.loc[comparison['Test MSE'].idxmin(), 'Model']
    print(f"\nBest Model (Lowest Test MSE): {best_model}")

# Compare models
compare_models([lin_train_mse, lin_test_mse, quad_train_mse, quad_test_mse, lasso_train_mse, lasso_test_mse])

# %% [markdown]
# ### VIF for Linear Regression
def check_vif(X_train, predictors=None):
    if predictors is None:
        predictors = X_train.columns
    X = X_train[predictors].fillna(X_train[predictors].median())
    X = (X - X.mean()) / X.std()
    vif_data = pd.DataFrame()
    vif_data['Predictor'] = predictors
    vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    print("\nVariance Inflation Factors (Linear Regression):")
    display(vif_data)
    return vif_data

# Check VIF for key predictors
vif_data = check_vif(X_train, predictors=['Log_GrLivArea', 'Log_LotArea', 'Log_TotalBsmtSF', 
                                         'TotalSF', 'Age', 'OverallQual', 'Qual_GrLivArea'])

# %% [markdown]
# ### Save Processed Data (Optional)
# df.to_csv('data/ames_processed.csv', index=False)
# print("Processed data saved to 'data/ames_processed.csv'")
```

# Step 4: Model Evaluation

```python
# %% [markdown]
# # ISOM5610: Introduction to Business Analytics
# ## Step 4: Model Evaluation and Selection
# This notebook evaluates the performance of Multiple Linear Regression, Quadratic Regression, and LASSO Regression models, selects the best model, and derives business insights for house price prediction in Ames, Iowa. Outputs are displayed inline.

# %% [markdown]
# ### Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.metrics import mean_squared_error, r2_score
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.stats.diagnostic import het_breuschpagan
from scipy.stats import shapiro
import statsmodels.api as sm
from IPython.display import display

# Set random seed for reproducibility
np.random.seed(42)

# Set plotting style
try:
    plt.style.use('ggplot')
except OSError:
    print("Warning: 'ggplot' style not found. Using default Matplotlib style.")
sns.set(font_scale=1.2)

# %% [markdown]
# ### Load and Preprocess Data (from Step 3)
def load_and_preprocess_data(file_path='data/Ames.csv'):
    try:
        df = pd.read_csv(file_path)
        print(f"Dataset loaded with {df.shape[0]} rows and {df.shape[1]} columns.")
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found. Please ensure 'Ames.csv' is in the 'data' subfolder.")
        raise
    
    # Handle missing values
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
    for col in numerical_cols:
        df[col] = df[col].fillna(df[col].median())
    none_cols = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 
                 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 
                 'BsmtFinType2', 'GarageType', 'GarageFinish', 'GarageQual', 
                 'GarageCond', 'MasVnrType']
    for col in none_cols:
        df[col] = df[col].fillna('None')
    df['Electrical'] = df['Electrical'].fillna(df['Electrical'].mode()[0])
    df = df.drop(columns=['PoolQC', 'MiscFeature', 'Alley', 'Fence'])
    
    # Encode categorical variables
    quality_map = {'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5, 'None': 0, 
                   np.nan: 0, 'Unknown': 0}
    ordinal_cols = ['ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond', 
                    'HeatingQC', 'KitchenQual', 'FireplaceQu', 'GarageQual', 
                    'GarageCond']
    for col in ordinal_cols:
        df[col] = df[col].map(quality_map, na_action='ignore').fillna(0).astype(float)
    
    categorical_cols = ['MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities', 
                        'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 
                        'Condition2', 'BldgType', 'HouseStyle', 'RoofStyle', 
                        'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 
                        'Foundation', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 
                        'Heating', 'CentralAir', 'Electrical', 'Functional', 
                        'GarageType', 'GarageFinish', 'PavedDrive', 'SaleType', 
                        'SaleCondition']
    df = pd.get_dummies(df, columns=categorical_cols, drop_first=True, dtype=float)
    print(f"Encoded categorical variables: {len(categorical_cols)} columns")
    
    # Log-transform skewed variables
    skewed_cols = ['SalePrice', 'GrLivArea', 'LotArea', 'TotalBsmtSF']
    for col in skewed_cols:
        df[f'Log_{col}'] = np.log1p(df[col]).astype(float)
    
    # Feature engineering
    df['TotalSF'] = (df['TotalBsmtSF'] + df['1stFlrSF'] + df['2ndFlrSF']).astype(float)
    df['Age'] = (df['YrSold'] - df['YearBuilt']).astype(float)
    df['Qual_GrLivArea'] = (df['OverallQual'] * df['Log_GrLivArea']).astype(float)
    print("Engineered features: TotalSF, Age, Qual_GrLivArea")
    
    # Standardize numerical predictors
    scaler = StandardScaler()
    num_cols = ['Log_GrLivArea', 'Log_LotArea', 'Log_TotalBsmtSF', 'TotalSF', 
                'Age', 'OverallQual', 'Qual_GrLivArea']
    df[num_cols] = scaler.fit_transform(df[num_cols]).astype(float)
    print("Standardized numerical predictors:", num_cols)
    
    # Verify all columns are numeric
    non_numeric_cols = df.select_dtypes(include=['object']).columns
    if len(non_numeric_cols) > 0:
        print(f"Warning: Non-numeric columns detected: {non_numeric_cols}")
        raise ValueError("Non-numeric columns found after preprocessing.")
    else:
        print("All columns are numeric.")
    
    # Display processed dataset summary
    print("\nProcessed Dataset Summary:")
    display(df[['Log_SalePrice', 'Log_GrLivArea', 'Log_LotArea', 'Log_TotalBsmtSF', 
                'TotalSF', 'Age', 'OverallQual', 'Qual_GrLivArea']].describe())
    
    return df

# Load and preprocess
df = load_and_preprocess_data()
print(f"Preprocessed dataset shape: {df.shape}")

# %% [markdown]
# ### Data Partition
def partition_data(df, target='Log_SalePrice'):
    features = [col for col in df.columns if col not in ['SalePrice', 'Log_SalePrice', 'Id']]
    X = df[features]
    y = df[target].astype(float)
    
    # Verify X is numeric
    if X.select_dtypes(include=['object']).columns.any():
        raise ValueError(f"Non-numeric columns in X: {X.select_dtypes(include=['object']).columns}")
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    print(f"Training set shape: {X_train.shape}, Testing set shape: {X_test.shape}")
    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = partition_data(df)

# %% [markdown]
# ### Model 1: Multiple Linear Regression
def fit_linear_regression(X_train, X_test, y_train, y_test):
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    train_mse = mean_squared_error(y_train, y_train_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)
    train_rmse = np.sqrt(train_mse)
    test_rmse = np.sqrt(test_mse)
    r2_train = r2_score(y_train, y_train_pred)
    r2_test = r2_score(y_test, y_test_pred)
    
    X_train_sm = sm.add_constant(X_train)
    model_sm = sm.OLS(y_train, X_train_sm).fit()
    
    residuals = y_train - y_train_pred
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    sns.scatterplot(x=y_train_pred, y=residuals)
    plt.axhline(0, color='red', linestyle='--')
    plt.title('Residuals vs Fitted (Linear Regression)')
    plt.xlabel('Fitted Values')
    plt.ylabel('Residuals')
    
    plt.subplot(1, 2, 2)
    sm.qqplot(residuals, line='45', fit=True)
    plt.title('Q-Q Plot (Linear Regression)')
    plt.tight_layout()
    plt.show()
    
    _, pval_bp, _, _ = het_breuschpagan(residuals, X_train_sm)
    _, pval_sw = shapiro(residuals)
    
    # Adjusted R-squared
    adj_r2_train = 1 - (1 - r2_train) * (len(y_train) - 1) / (len(y_train) - X_train.shape[1] - 1)
    adj_r2_test = 1 - (1 - r2_test) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1)
    
    # Cross-validation
    cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
    cv_rmse = np.sqrt(-cv_scores.mean())
    
    print("\nLinear Regression Results:")
    print(f"Train MSE: {train_mse:.4f}, Train RMSE: {train_rmse:.4f}")
    print(f"Test MSE: {test_mse:.4f}, Test RMSE: {test_rmse:.4f}")
    print(f"Train Adjusted R-squared: {adj_r2_train:.4f}, Test Adjusted R-squared: {adj_r2_test:.4f}")
    print(f"5-Fold CV RMSE: {cv_rmse:.4f}")
    print(f"Breusch-Pagan p-value: {pval_bp:.4f} (H0: Homoskedasticity)")
    print(f"Shapiro-Wilk p-value: {pval_sw:.4f} (H0: Normality)")
    print("\nModel Summary:")
    display(model_sm.summary())
    
    return model, model_sm, train_mse, test_mse, adj_r2_train, adj_r2_test, cv_rmse

# Fit linear regression
lin_reg, lin_reg_sm, lin_train_mse, lin_test_mse, lin_adj_r2_train, lin_adj_r2_test, lin_cv_rmse = fit_linear_regression(X_train, X_test, y_train, y_test)

# %% [markdown]
# ### Model 2: Quadratic Regression
def fit_quadratic_regression(X_train, X_test, y_train, y_test):
    X_train_quad = X_train.copy()
    X_test_quad = X_test.copy()
    quad_cols = ['Log_GrLivArea', 'Log_TotalBsmtSF']
    for col in quad_cols:
        X_train_quad[f'{col}_squared'] = X_train_quad[col] ** 2
        X_test_quad[f'{col}_squared'] = X_test_quad[col] ** 2
    
    model = LinearRegression()
    model.fit(X_train_quad, y_train)
    
    y_train_pred = model.predict(X_train_quad)
    y_test_pred = model.predict(X_test_quad)
    
    train_mse = mean_squared_error(y_train, y_train_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)
    train_rmse = np.sqrt(train_mse)
    test_rmse = np.sqrt(test_mse)
    r2_train = r2_score(y_train, y_train_pred)
    r2_test = r2_score(y_test, y_test_pred)
    
    residuals = y_train - y_train_pred
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    sns.scatterplot(x=y_train_pred, y=residuals)
    plt.axhline(0, color='red', linestyle='--')
    plt.title('Residuals vs Fitted (Quadratic Regression)')
    plt.xlabel('Fitted Values')
    plt.ylabel('Residuals')
    
    plt.subplot(1, 2, 2)
    sm.qqplot(residuals, line='45', fit=True)
    plt.title('Q-Q Plot (Quadratic Regression)')
    plt.tight_layout()
    plt.show()
    
    X_train_sm = sm.add_constant(X_train_quad)
    _, pval_bp, _, _ = het_breuschpagan(residuals, X_train_sm)
    _, pval_sw = shapiro(residuals)
    
    adj_r2_train = 1 - (1 - r2_train) * (len(y_train) - 1) / (len(y_train) - X_train_quad.shape[1] - 1)
    adj_r2_test = 1 - (1 - r2_test) * (len(y_test) - 1) / (len(y_test) - X_test_quad.shape[1] - 1)
    
    cv_scores = cross_val_score(model, X_train_quad, y_train, cv=5, scoring='neg_mean_squared_error')
    cv_rmse = np.sqrt(-cv_scores.mean())
    
    print("\nQuadratic Regression Results:")
    print(f"Train MSE: {train_mse:.4f}, Train RMSE: {train_rmse:.4f}")
    print(f"Test MSE: {test_mse:.4f}, Test RMSE: {test_rmse:.4f}")
    print(f"Train Adjusted R-squared: {adj_r2_train:.4f}, Test Adjusted R-squared: {adj_r2_test:.4f}")
    print(f"5-Fold CV RMSE: {cv_rmse:.4f}")
    print(f"Breusch-Pagan p-value: {pval_bp:.4f} (H0: Homoskedasticity)")
    print(f"Shapiro-Wilk p-value: {pval_sw:.4f} (H0: Normality)")
    
    return model, train_mse, test_mse, adj_r2_train, adj_r2_test, cv_rmse

# Fit quadratic regression
quad_reg, quad_train_mse, quad_test_mse, quad_adj_r2_train, quad_adj_r2_test, quad_cv_rmse = fit_quadratic_regression(X_train, X_test, y_train, y_test)

# %% [markdown]
# ### Model 3: LASSO Regression
def fit_lasso_regression(X_train, X_test, y_train, y_test):
    alphas = np.logspace(-4, 1, 50)
    best_alpha = None
    best_score = -np.inf
    for alpha in alphas:
        model = Lasso(alpha=alpha, max_iter=10000)
        scores = cross_val_score(model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
        if scores.mean() > best_score:
            best_score = scores.mean()
            best_alpha = alpha
    
    model = Lasso(alpha=best_alpha, max_iter=10000)
    model.fit(X_train, y_train)
    
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    train_mse = mean_squared_error(y_train, y_train_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)
    train_rmse = np.sqrt(train_mse)
    test_rmse = np.sqrt(test_mse)
    r2_train = r2_score(y_train, y_train_pred)
    r2_test = r2_score(y_test, y_test_pred)
    
    residuals = y_train - y_train_pred
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    sns.scatterplot(x=y_train_pred, y=residuals)
    plt.axhline(0, color='red', linestyle='--')
    plt.title('Residuals vs Fitted (LASSO Regression)')
    plt.xlabel('Fitted Values')
    plt.ylabel('Residuals')
    
    plt.subplot(1, 2, 2)
    sm.qqplot(residuals, line='45', fit=True)
    plt.title('Q-Q Plot (LASSO Regression)')
    plt.tight_layout()
    plt.show()
    
    X_train_sm = sm.add_constant(X_train)
    _, pval_bp, _, _ = het_breuschpagan(residuals, X_train_sm)
    _, pval_sw = shapiro(residuals)
    
    selected_features = X_train.columns[model.coef_ != 0].tolist()
    coef_df = pd.DataFrame({
        'Feature': X_train.columns,
        'Coefficient': model.coef_
    }).sort_values(by='Coefficient', key=abs, ascending=False)
    
    adj_r2_train = 1 - (1 - r2_train) * (len(y_train) - 1) / (len(y_train) - X_train.shape[1] - 1)
    adj_r2_test = 1 - (1 - r2_test) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1)
    
    cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
    cv_rmse = np.sqrt(-cv_scores.mean())
    
    print("\nLASSO Regression Results:")
    print(f"Best alpha: {best_alpha:.6f}")
    print(f"Train MSE: {train_mse:.4f}, Train RMSE: {train_rmse:.4f}")
    print(f"Test MSE: {test_mse:.4f}, Test RMSE: {test_rmse:.4f}")
    print(f"Train Adjusted R-squared: {adj_r2_train:.4f}, Test Adjusted R-squared: {adj_r2_test:.4f}")
    print(f"5-Fold CV RMSE: {cv_rmse:.4f}")
    print(f"Breusch-Pagan p-value: {pval_bp:.4f} (H0: Homoskedasticity)")
    print(f"Shapiro-Wilk p-value: {pval_sw:.4f} (H0: Normality)")
    print(f"Selected features ({len(selected_features)}): {selected_features}")
    print("\nTop Coefficients:")
    display(coef_df.head(10))
    
    return model, train_mse, test_mse, adj_r2_train, adj_r2_test, cv_rmse, selected_features, coef_df

# Fit LASSO regression
lasso_reg, lasso_train_mse, lasso_test_mse, lasso_adj_r2_train, lasso_adj_r2_test, lasso_cv_rmse, lasso_selected_features, lasso_coef_df = fit_lasso_regression(X_train, X_test, y_train, y_test)

# %% [markdown]
# ### Model Comparison and Selection
def compare_models():
    comparison = pd.DataFrame({
        'Model': ['Linear Regression', 'Quadratic Regression', 'LASSO Regression'],
        'Train MSE': [lin_train_mse, quad_train_mse, lasso_train_mse],
        'Test MSE': [lin_test_mse, quad_test_mse, lasso_test_mse],
        'Train RMSE': [np.sqrt(lin_train_mse), np.sqrt(quad_train_mse), np.sqrt(lasso_train_mse)],
        'Test RMSE': [np.sqrt(lin_test_mse), np.sqrt(quad_test_mse), np.sqrt(lasso_test_mse)],
        'Train Adjusted R-squared': [lin_adj_r2_train, quad_adj_r2_train, lasso_adj_r2_train],
        'Test Adjusted R-squared': [lin_adj_r2_test, quad_adj_r2_test, lasso_adj_r2_test],
        '5-Fold CV RMSE': [lin_cv_rmse, quad_cv_rmse, lasso_cv_rmse]
    })
    print("\nModel Comparison:")
    display(comparison)
    
    best_model = comparison.loc[comparison['Test MSE'].idxmin(), 'Model']
    print(f"\nBest Model (Lowest Test MSE): {best_model}")
    
    return comparison, best_model

# Compare models
comparison, best_model = compare_models()

# %% [markdown]
# ### VIF for Linear Regression
def check_vif(X_train, predictors=None):
    if predictors is None:
        predictors = ['Log_GrLivArea', 'Log_LotArea', 'Log_TotalBsmtSF', 'TotalSF', 
                      'Age', 'OverallQual', 'Qual_GrLivArea']
    X = X_train[predictors].fillna(X_train[predictors].median())
    X = (X - X.mean()) / X.std()
    vif_data = pd.DataFrame()
    vif_data['Predictor'] = predictors
    vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    print("\nVariance Inflation Factors (Linear Regression):")
    display(vif_data)
    return vif_data

# Check VIF
vif_data = check_vif(X_train)

# %% [markdown]
# ### Interpretation and Business Insights
def interpret_results(coef_df, selected_features):
    print("\nInterpretation of LASSO Model:")
    print("The LASSO model coefficients represent the percentage change in SalePrice for a one-unit increase in standardized predictors (since Log_SalePrice is the target).")
    print("\nKey Findings:")
    top_features = coef_df[coef_df['Feature'].isin(selected_features)].head(5)
    for _, row in top_features.iterrows():
        feature, coef = row['Feature'], row['Coefficient']
        print(f"- {feature}: A 1-standard-deviation increase is associated with a {coef*100:.2f}% change in SalePrice.")
    
    print("\nBusiness Insights:")
    print("- **Property Selection**: Target properties with high OverallQual, large GrLivArea, and located in premium neighborhoods (e.g., NoRidge, NridgHt) for higher resale value.")
    print("- **Renovation Priorities**: Focus on improving KitchenQual, ExterQual, and adding livable space (e.g., finishing basements to increase TotalSF), as these significantly increase SalePrice.")
    print("- **Price Estimation**: Use the LASSO model to predict SalePrice for potential purchases, ensuring accurate pricing for profit optimization.")
    print("- **Unexpected Findings**: Negative coefficients for some Neighborhoods may indicate omitted variable bias (Topic 2b). Consider additional features (e.g., school quality) in future analyses.")
    
    print("\nTop Coefficients for Report:")
    display(top_features)

# Interpret results
interpret_results(lasso_coef_df, lasso_selected_features)

# %% [markdown]
# ### Save Processed Data (Optional)
# df.to_csv('data/ames_processed.csv', index=False)
# print("Processed data saved to 'data/ames_processed.csv'")
```
