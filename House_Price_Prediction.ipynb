```python
# %% [markdown]
# # ISOM5610: Introduction to Business Analytics
# ## Step 1: Data Understanding and Exploratory Data Analysis (EDA)
# This notebook performs EDA on the Ames house price dataset to understand its structure, identify patterns, and detect issues (e.g., missing values, outliers, multicollinearity). Outputs are displayed inline for review.

# %% [markdown]
# ### Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.outliers_influence import variance_inflation_factor
from IPython.display import display

# Set up plotting style
try:
    plt.style.use('ggplot')  # Use 'ggplot' style for clean aesthetics
except OSError:
    print("Warning: 'ggplot' style not found. Using default Matplotlib style.")
sns.set(font_scale=1.2)

# %% [markdown]
# ### Step 1.1: Load the Dataset
def load_data(file_path='data/Ames.csv'):
    try:
        df = pd.read_csv(file_path)
        print(f"Dataset loaded with {df.shape[0]} rows and {df.shape[1]} columns.")
        return df
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found. Please ensure 'Ames.csv' is in the 'data' subfolder.")
        raise

# Load dataset
df = load_data()

# %% [markdown]
# ### Step 1.2: Summarize Data
def summarize_data(df):
    # Descriptive statistics for numerical variables
    numerical_summary = df.select_dtypes(include=['int64', 'float64']).describe()
    print("\nDescriptive Statistics for Numerical Variables:")
    display(numerical_summary)
    
    # Summary for categorical variables
    categorical_cols = df.select_dtypes(include=['object']).columns
    print("\nCategorical Variables Summary:")
    for col in categorical_cols:
        print(f"{col}: {df[col].nunique()} unique values, Mode: {df[col].mode()[0]}")
    
    return numerical_summary, categorical_cols

# Summarize data
numerical_summary, categorical_cols = summarize_data(df)

# %% [markdown]
# ### Step 1.3: Visualize Distributions
def plot_distributions(df, key_numerical=['SalePrice', 'GrLivArea', 'LotArea', 'TotalBsmtSF'], 
                       key_categorical=['Neighborhood', 'OverallQual']):
    # Numerical variables: Histograms and Boxplots
    for col in key_numerical:
        plt.figure(figsize=(12, 5))
        
        # Histogram
        plt.subplot(1, 2, 1)
        sns.histplot(df[col], kde=True)
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        
        # Boxplot
        plt.subplot(1, 2, 2)
        sns.boxplot(y=df[col])
        plt.title(f'Boxplot of {col}')
        
        plt.tight_layout()
        plt.show()
    
    # Categorical variables: Boxplots of SalePrice
    for col in key_categorical:
        plt.figure(figsize=(10, 6))
        sns.boxplot(x=col, y='SalePrice', data=df)
        plt.title(f'SalePrice by {col}')
        plt.xticks(rotation=45)
        plt.show()

# Plot distributions
plot_distributions(df)

# %% [markdown]
# ### Step 1.4: Explore Relationships
def explore_relationships(df, predictors=['GrLivArea', 'LotArea', 'TotalBsmtSF']):
    # Scatterplots
    for col in predictors:
        plt.figure(figsize=(8, 6))
        sns.scatterplot(x=df[col], y=df['SalePrice'])
        plt.title(f'SalePrice vs {col}')
        plt.xlabel(col)
        plt.ylabel('SalePrice')
        plt.show()
    
    # Correlation matrix
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
    corr_matrix = df[numerical_cols].corr()
    plt.figure(figsize=(12, 10))
    sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', center=0)
    plt.title('Correlation Matrix')
    plt.show()
    
    # Top correlations with SalePrice
    saleprice_corr = corr_matrix['SalePrice'].sort_values(ascending=False)
    print("\nTop Correlations with SalePrice:")
    display(saleprice_corr.head(10))
    return corr_matrix

# Explore relationships
corr_matrix = explore_relationships(df)

# %% [markdown]
# ### Step 1.5: Check for Multicollinearity
def check_multicollinearity(df, predictors=['GrLivArea', 'LotArea', 'TotalBsmtSF', 'OverallQual', 
                                            'YearBuilt', '1stFlrSF']):
    # Prepare data for VIF
    X = df[predictors].fillna(df[predictors].median())
    X = (X - X.mean()) / X.std()  # Standardize to avoid scale issues
    vif_data = pd.DataFrame()
    vif_data['Predictor'] = predictors
    vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    print("\nVariance Inflation Factors:")
    display(vif_data)
    return vif_data

# Check multicollinearity
vif_data = check_multicollinearity(df)

# %% [markdown]
# ### Step 1.6: Assess Missing Values
def assess_missing_values(df):
    missing = df.isnull().sum()
    missing_percent = (missing / len(df)) * 100
    missing_df = pd.DataFrame({'Missing Count': missing, 'Missing Percent': missing_percent})
    missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values(by='Missing Count', ascending=False)
    print("\nMissing Values:")
    display(missing_df)
    return missing_df

# Assess missing values
missing_df = assess_missing_values(df)

# %% [markdown]
# ### Step 1.7: Detect Outliers
def detect_outliers(df, columns=['SalePrice', 'GrLivArea', 'LotArea']):
    outliers = {}
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outlier_count = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
        outliers[col] = outlier_count
        print(f"\nOutliers in {col}: {outlier_count}")
    return outliers

# Detect outliers
outliers = detect_outliers(df)
```
